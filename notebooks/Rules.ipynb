```python
# Business Rules Class
class BusinessRules:
    def __init__(self, df):
        self.df = df
        # Convert Date_Time to datetime format once for efficiency
        self.df['Date_Time'] = pd.to_datetime(self.df['Date_Time'], format='%Y-%m-%d %H:%M:%S')

    def rule1_n_transactions_m_minutes(self, max_transactions=3, max_minutes=30):
        flagged_transactions = []
        for _, row in self.df.iterrows():
            date_time = row['Date_Time']
            recent_transactions = self.df[(self.df['Date_Time'] <= date_time) & 
                                          (self.df['Date_Time'] >= date_time - timedelta(minutes=max_minutes))]
            if len(recent_transactions) > max_transactions:
                flagged_transactions.append(row['Transaction_Id'])
        return flagged_transactions

    def rule2_high_risk_cities(self, high_risk_cities=['Sindh', 'Balochistan']):
        flagged_transactions = self.df[self.df['Dest_State'].isin(high_risk_cities)]
        return flagged_transactions['Transaction_Id'].tolist()

    def rule3_high_risk_region_and_time(self, high_risk_region='Balochistan', high_risk_times=['03:00', '04:00']):
        flagged_transactions = []
        for _, row in self.df.iterrows():
            # Extract time part for comparison
            time = row['Date_Time'].strftime('%H:%M')
            if row['Dest_State'] == high_risk_region and time in high_risk_times:
                flagged_transactions.append(row['Transaction_Id'])
        return flagged_transactions

    def rule4_new_account_with_large_transaction(self, amount_threshold=50000, days_threshold=30):
        # Ensure 'Account_Creation_Date' is in datetime format
        self.df['Account_Creation_Date'] = pd.to_datetime(self.df['Account_Creation_Date'], errors='coerce')

        # Calculate the cutoff date for "new" accounts (created within the last 30 days)
        cutoff_date = datetime.now() - timedelta(days=days_threshold)

        # Apply Rule 4: Flag transactions where the receiving account is newly created and the amount exceeds the threshold
        self.df['R4_Flag_NewAccount_Amount'] = (
            (self.df['Account_Creation_Date'] >= cutoff_date) &  # Account is new
            (self.df['Amount'] > amount_threshold)  # Amount exceeds threshold
        )

        # Get the flagged transactions for Rule 4
        flagged_r4 = self.df[self.df['R4_Flag_NewAccount_Amount'] == True]
        return flagged_r4['Transaction_Id'].tolist()

df = pd.DataFrame(data)

# Initialize Business Rules Class
business_rules = BusinessRules(df)

# Apply Business Rules
rule1_results = business_rules.rule1_n_transactions_m_minutes()
rule2_results = business_rules.rule2_high_risk_cities()
rule3_results = business_rules.rule3_high_risk_region_and_time()
rule4_results = business_rules.rule4_new_account_with_large_transaction()

# Combine flagged transactions from all rules
flagged_transactions = set(rule1_results + rule2_results + rule3_results + rule4_results)

# Flag the transactions in the dataframe
df['Flagged'] = df['Transaction_Id'].isin(flagged_transactions)

# Separate flagged and non-flagged transactions
flagged = df[df['Flagged']]
non_flagged = df[~df['Flagged']]

# Print results
print(f"üö® Flagged Transactions: {len(flagged)}")
print(f"‚úÖ Non-Flagged Transactions: {len(non_flagged)}\n")
print("üîç Flagged Transaction Details:")
print(flagged[['Transaction_Id', 'Date_Time', 'Account_Creation_Date', 'Phone_Number',
       'CNIC', 'Name', 'Type', 'Amount', 'ID_Source', 'Old_Balance',
       'New_Balance', 'Device_Name', 'IMEI', 'KYC_Status', 'Service_Charges',
       'Channel', 'Remarks', 'ID_Dest', 'Dest_State', 'Dest_City', 'Is_Fraud']])
```
