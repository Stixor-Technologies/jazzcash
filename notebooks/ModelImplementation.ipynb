```python
def preprocess(df):
    df['Date Time'] = pd.to_datetime(df['Date Time'])
    df['year'] = df['Date Time'].dt.year
    df['month'] = df['Date Time'].dt.month
    df['day'] = df['Date Time'].dt.day
    df['hour'] = df['Date Time'].dt.hour
    df['minute'] = df['Date Time'].dt.minute
    df['second'] = df['Date Time'].dt.second
    df['weekday'] = df['Date Time'].dt.weekday
    df['day_of_year'] = df['Date Time'].dt.dayofyear

    df = df.drop(['Transaction Id', 'Date Time', 'Phone Number', 'CNIC', 'Name'],axis = 1)

    objs = []
    for col in df.columns:
        if df[col].dtype == 'O':
            objs.append(col)

    objs.remove('Remarks')
    
    cont_features = ['Amount', 'Old Balance', 'New Balance','Service Charges']
    
    
    df_normalized = df.copy()
    #scaler = MinMaxScaler()
    
    #df_normalized[cont_features] = scaler.fit_transform(df[cont_features])
    
    

    # label_encoder object knows how to understand word labels.
    label_encoder = preprocessing.LabelEncoder()
    le_data=df_normalized

    for col in objs:
        le_data[col]= label_encoder.fit_transform(le_data[col])
 
    le_data = le_data.drop(['Remarks'],axis = 1)

    return le_data

    
```

```python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd
from sklearn.utils import resample

# Step 1: Check if x_train is a 1D array and reshape it if needed
if len(x_train.shape) == 1:
    x_train = x_train.reshape(-1, 1)  # Reshape x_train into 2D (n_samples, 1)

# Step 2: Apply RandomUnderSampler to balance the classes
rus = RandomUnderSampler(sampling_strategy=1.0, random_state=42)  # Equal sampling ratio
X_train_resampled, y_train_resampled = rus.fit_resample(x_train, y_train)

# Step 3: Downsample y_train_resampled to match y_train shape
X_train_resampled, y_train_resampled = resample(X_train_resampled, y_train_resampled, 
                                                n_samples=len(y_train), 
                                                random_state=42)

# Step 4: Verify the shapes
print(f"y_train shape: {y_train.shape}")
print(f"y_train_resampled shape: {y_train_resampled.shape}")
```

```python
from sklearn.utils import resample
import pandas as pd

# Step 1: Reshape x_train if it's 1D
if len(x_train.shape) == 1:
    x_train = x_train.reshape(-1, 1)

# Step 2: Separate the classes
x_train_class_0 = x_train[y_train == 0]
y_train_class_0 = y_train[y_train == 0]

x_train_class_1 = x_train[y_train == 1]
y_train_class_1 = y_train[y_train == 1]

# Step 3: Downsample class 0 to half its size
half_class_0_size = int(len(y_train_class_0) / 2)  # or use round(...) if preferred
x_train_class_0_resampled, y_train_class_0_resampled = resample(
    x_train_class_0,
    y_train_class_0,
    n_samples=half_class_0_size,
    random_state=42
)

# Step 4: Combine the downsampled class 0 with the full class 1
X_train_resampled = pd.concat([
    pd.DataFrame(x_train_class_0_resampled),
    pd.DataFrame(x_train_class_1)
], axis=0)

y_train_resampled = pd.concat([
    pd.Series(y_train_class_0_resampled),
    pd.Series(y_train_class_1)
], axis=0)



# Step 5: Shuffle the dataset
X_train_resampled, y_train_resampled = resample(
    X_train_resampled,
    y_train_resampled,
    n_samples=len(X_train_resampled),
    random_state=42
)

# Step 6: Print results
print(f"Original y_train distribution:\n{y_train.value_counts()}")
print(f"Resampled y_train distribution:\n{y_train_resampled.value_counts()}")
```

```python
# Step 4: Verify the shapes
print(f"y_train shape: {y_train.shape}")
print(f"y_train_resampled shape: {y_train_resampled.shape}")
```
