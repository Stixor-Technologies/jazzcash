```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set up visualization style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Load all three datasets
try:
    model_data = pd.read_csv('/home/nazar/Muhammad Raees Azam/JazzCash/Datasets/Dataset_model.csv')
    original_data = pd.read_csv('/home/nazar/Muhammad Raees Azam/JazzCash/Datasets/Synthetic_Financial_datasets_log_Original.csv')
    generated_data = pd.read_csv('/home/nazar/Muhammad Raees Azam/JazzCash/Datasets/SyntheticTransactions_Generated.csv')
    
    print("All datasets loaded successfully!")
except Exception as e:
    print(f"Error loading datasets: {e}")

## 1. Basic Dataset Exploration

def basic_eda(df, name):
    print(f"\n=== {name} Dataset Basic Info ===")
    print(f"Shape: {df.shape}")
    print("\nFirst 5 rows:")
    display(df.head())
    print("\nData Types:")
    print(df.dtypes)
    print("\nMissing Values:")
    print(df.isnull().sum())
    print("\nDescriptive Statistics:")
    display(df.describe(include='all'))

# Perform basic EDA for each dataset
basic_eda(model_data, "Original Dataset 2")
basic_eda(original_data, "Original Dataset 1")
basic_eda(generated_data, "Generated")

## 2. Comparative Analysis of Numerical Features

def compare_numerical_features(original, generated, features):
    fig, axes = plt.subplots(len(features), 2, figsize=(15, 5*len(features)))
    
    for i, feature in enumerate(features):
        # Original data distribution
        sns.histplot(original[feature], ax=axes[i,0], kde=True, color='blue')
        axes[i,0].set_title(f'Original {feature} Distribution')
        axes[i,0].set_xlabel(feature)
        
        # Generated data distribution
        sns.histplot(generated[feature], ax=axes[i,1], kde=True, color='green')
        axes[i,1].set_title(f'Generated {feature} Distribution')
        axes[i,1].set_xlabel(feature)
    
    plt.tight_layout()
    plt.show()

# Identify common numerical features (adjust based on your actual columns)
common_numerical = ['amount', 'step']  # Replace with your actual numerical columns
compare_numerical_features(original_data, generated_data, common_numerical)

## 3. Temporal Pattern Analysis

def analyze_temporal_patterns(original, generated, time_col='timestamp'):
    # Convert timestamps if needed
    if not pd.api.types.is_datetime64_any_dtype(original[time_col]):
        original[time_col] = pd.to_datetime(original[time_col])
    if not pd.api.types.is_datetime64_any_dtype(generated[time_col]):
        generated[time_col] = pd.to_datetime(generated[time_col])
    
    # Extract temporal features
    for df in [original, generated]:
        df['hour'] = df[time_col].dt.hour
        df['day_of_week'] = df[time_col].dt.dayofweek
        df['day_of_month'] = df[time_col].dt.day
    
    # Plot comparisons
    fig, axes = plt.subplots(3, 2, figsize=(15, 12))
    
    # Hourly distribution
    sns.countplot(x='hour', data=original, ax=axes[0,0], color='blue')
    axes[0,0].set_title('Original - Hourly Distribution')
    sns.countplot(x='hour', data=generated, ax=axes[0,1], color='green')
    axes[0,1].set_title('Generated - Hourly Distribution')
    
    # Day of week distribution
    sns.countplot(x='day_of_week', data=original, ax=axes[1,0], color='blue')
    axes[1,0].set_title('Original - Day of Week Distribution')
    sns.countplot(x='day_of_week', data=generated, ax=axes[1,1], color='green')
    axes[1,1].set_title('Generated - Day of Week Distribution')
    
    # Day of month distribution
    sns.countplot(x='day_of_month', data=original, ax=axes[2,0], color='blue')
    axes[2,0].set_title('Original - Day of Month Distribution')
    sns.countplot(x='day_of_month', data=generated, ax=axes[2,1], color='green')
    axes[2,1].set_title('Generated - Day of Month Distribution')
    
    plt.tight_layout()
    plt.show()

# Analyze temporal patterns if timestamp exists
if 'timestamp' in original_data.columns and 'timestamp' in generated_data.columns:
    analyze_temporal_patterns(original_data, generated_data)

## 4. Categorical Feature Comparison

def compare_categorical_features(original, generated, features):
    fig, axes = plt.subplots(len(features), 2, figsize=(15, 5*len(features)))
    
    for i, feature in enumerate(features):
        # Original data distribution
        value_counts_orig = original[feature].value_counts(normalize=True)
        value_counts_orig.plot(kind='bar', ax=axes[i,0], color='blue')
        axes[i,0].set_title(f'Original {feature} Distribution')
        axes[i,0].set_xlabel(feature)
        
        # Generated data distribution
        value_counts_gen = generated[feature].value_counts(normalize=True)
        value_counts_gen.plot(kind='bar', ax=axes[i,1], color='green')
        axes[i,1].set_title(f'Generated {feature} Distribution')
        axes[i,1].set_xlabel(feature)
    
    plt.tight_layout()
    plt.show()

# Identify common categorical features (adjust based on your actual columns)
common_categorical = ['type', 'isFraud']  # Replace with your actual categorical columns
compare_categorical_features(original_data, generated_data, common_categorical)

## 5. Model Dataset Analysis (Rules/Patterns)

def analyze_model_dataset(model_data):
    print("\n=== Model Dataset Analysis ===")
    
    # Check if model contains distribution parameters
    if 'distribution' in model_data.columns:
        print("\nDistribution Parameters:")
        display(model_data[['feature', 'distribution', 'parameters']].drop_duplicates())
    
    # Check for behavioral patterns
    if 'pattern_type' in model_data.columns:
        print("\nBehavioral Patterns:")
        display(model_data[['pattern_type', 'description', 'parameters']].drop_duplicates())
    
    # Check for fraud patterns if exists
    if 'fraud_pattern' in model_data.columns:
        print("\nFraud Patterns:")
        display(model_data[model_data['fraud_pattern'] == True])
    
    # Visualize any numerical parameters in model
    numerical_cols = model_data.select_dtypes(include=np.number).columns
    if len(numerical_cols) > 0:
        print("\nNumerical Parameters Distribution:")
        model_data[numerical_cols].hist(figsize=(12, 8))
        plt.tight_layout()
        plt.show()

analyze_model_dataset(model_data)

## 6. Generated Dataset Validation

def validate_generated_data(model_data, original_data, generated_data):
    print("\n=== Generated Data Validation ===")
    
    # Check if generated data follows model distributions
    if 'amount' in generated_data.columns and 'amount' in original_data.columns:
        print("\nAmount Statistics Comparison:")
        comparison = pd.DataFrame({
            'Original': original_data['amount'].describe(),
            'Generated': generated_data['amount'].describe()
        })
        display(comparison)
        
        # KS test for distribution similarity
        from scipy.stats import ks_2samp
        ks_stat, p_value = ks_2samp(original_data['amount'], generated_data['amount'])
        print(f"\nKS Test for Amount Distribution: stat={ks_stat:.4f}, p-value={p_value:.4f}")
    
    # Check if fraud patterns match model specifications
    if 'isFraud' in generated_data.columns and 'fraud_ratio' in model_data.columns:
        expected_fraud_ratio = model_data['fraud_ratio'].mean()
        actual_fraud_ratio = generated_data['isFraud'].mean()
        print(f"\nFraud Ratio: Expected={expected_fraud_ratio:.4f}, Actual={actual_fraud_ratio:.4f}")
    
    # Check temporal patterns
    if 'timestamp' in generated_data.columns:
        print("\nTime Range Coverage:")
        print(f"Original: {original_data['timestamp'].min()} to {original_data['timestamp'].max()}")
        print(f"Generated: {generated_data['timestamp'].min()} to {generated_data['timestamp'].max()}")

validate_generated_data(model_data, original_data, generated_data)

## 7. Feature Correlation Analysis

def compare_correlations(original, generated, features):
    if len(features) < 2:
        print("Not enough features for correlation analysis")
        return
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # Original data correlation
    corr_orig = original[features].corr()
    sns.heatmap(corr_orig, annot=True, fmt=".2f", cmap='coolwarm', ax=ax1)
    ax1.set_title('Original Data Feature Correlation')
    
    # Generated data correlation
    corr_gen = generated[features].corr()
    sns.heatmap(corr_gen, annot=True, fmt=".2f", cmap='coolwarm', ax=ax2)
    ax2.set_title('Generated Data Feature Correlation')
    
    plt.tight_layout()
    plt.show()

# Select features for correlation analysis (adjust based on your data)
correlation_features = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'isFraud']  # Replace with your features
if all(feat in original_data.columns for feat in correlation_features) and \
   all(feat in generated_data.columns for feat in correlation_features):
    compare_correlations(original_data, generated_data, correlation_features)

print("\nEDA Complete! Review the visualizations and statistics above.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set up visualization style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Load all three datasets
try:
    model_data = pd.read_csv('Dataset_model.csv')
    original_data = pd.read_csv('Synthetic_Financial_datasets_log_Original.csv')
    generated_data = pd.read_csv('SyntheticTransactions_Generated.csv')
    
    print("All datasets loaded successfully!")
except Exception as e:
    print(f"Error loading datasets: {e}")

## 1. Enhanced Basic Dataset Exploration

def enhanced_eda(df, name):
    print(f"\n=== {name} Dataset Enhanced Analysis ===")
    print(f"Shape: {df.shape}")
    
    # Basic info
    print("\nFirst 3 rows:")
    display(df.head(3))
    
    # Data types and missing values
    print("\nData Types and Missing Values:")
    info_df = pd.DataFrame({
        'Data Type': df.dtypes,
        'Missing Values': df.isnull().sum(),
        'Unique Values': df.nunique()
    })
    display(info_df)
    
    # Numerical statistics
    print("\nNumerical Features Statistics:")
    display(df.describe(include=[np.number]))
    
    # Categorical statistics
    cat_cols = df.select_dtypes(include=['object', 'category']).columns
    if len(cat_cols) > 0:
        print("\nCategorical Features Statistics:")
        display(df[cat_cols].describe())

# Perform enhanced EDA for each dataset
enhanced_eda(original_data, "Original")
enhanced_eda(generated_data, "Generated")
if not model_data.empty:
    enhanced_eda(model_data, "Model")

## 2. Transaction Amount Analysis

def analyze_amounts(original, generated):
    if 'amt' not in original.columns or 'amt' not in generated.columns:
        print("'amt' column not found in both datasets")
        return
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    # Original amount distribution
    sns.histplot(original['amt'], ax=axes[0], kde=True, color='blue')
    axes[0].set_title('Original Amount Distribution')
    axes[0].set_xlabel('Transaction Amount')
    
    # Generated amount distribution
    sns.histplot(generated['amt'], ax=axes[1], kde=True, color='green')
    axes[1].set_title('Generated Amount Distribution')
    axes[1].set_xlabel('Transaction Amount')
    
    # Boxplot comparison
    combined = pd.concat([
        original['amt'].rename('Amount').to_frame().assign(Source='Original'),
        generated['amt'].rename('Amount').to_frame().assign(Source='Generated')
    ])
    sns.boxplot(x='Source', y='Amount', data=combined, ax=axes[2])
    axes[2].set_title('Amount Distribution Comparison')
    
    plt.tight_layout()
    plt.show()
    
    # Statistical comparison
    print("\nAmount Statistics Comparison:")
    comparison = pd.DataFrame({
        'Original': original['amt'].describe(),
        'Generated': generated['amt'].describe()
    })
    display(comparison)

analyze_amounts(original_data, generated_data)

## 3. Fraud Analysis

def analyze_fraud(original, generated):
    if 'is_fraud' not in original.columns or 'is_fraud' not in generated.columns:
        print("'is_fraud' column not found in both datasets")
        return
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Fraud distribution original
    fraud_counts_orig = original['is_fraud'].value_counts(normalize=True)
    fraud_counts_orig.plot(kind='bar', ax=axes[0], color=['blue', 'red'])
    axes[0].set_title('Original Fraud Distribution')
    axes[0].set_xlabel('Is Fraud')
    axes[0].set_ylabel('Proportion')
    
    # Fraud distribution generated
    fraud_counts_gen = generated['is_fraud'].value_counts(normalize=True)
    fraud_counts_gen.plot(kind='bar', ax=axes[1], color=['green', 'red'])
    axes[1].set_title('Generated Fraud Distribution')
    axes[1].set_xlabel('Is Fraud')
    axes[1].set_ylabel('Proportion')
    
    plt.tight_layout()
    plt.show()
    
    # Fraud rate comparison
    print("\nFraud Rate Comparison:")
    print(f"Original fraud rate: {original['is_fraud'].mean():.4f}")
    print(f"Generated fraud rate: {generated['is_fraud'].mean():.4f}")

analyze_fraud(original_data, generated_data)

## 4. Geographic Analysis

def analyze_geography(original, generated):
    geo_cols = ['lat', 'long', 'merch_lat', 'merch_long']
    if not all(col in original.columns for col in geo_cols) or not all(col in generated.columns for col in geo_cols):
        print("Geographic columns not found in both datasets")
        return
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Customer location comparison
    sns.scatterplot(x='long', y='lat', data=original, ax=axes[0,0], alpha=0.1, color='blue')
    axes[0,0].set_title('Original Customer Locations')
    sns.scatterplot(x='long', y='lat', data=generated, ax=axes[0,1], alpha=0.1, color='green')
    axes[0,1].set_title('Generated Customer Locations')
    
    # Merchant location comparison
    sns.scatterplot(x='merch_long', y='merch_lat', data=original, ax=axes[1,0], alpha=0.1, color='blue')
    axes[1,0].set_title('Original Merchant Locations')
    sns.scatterplot(x='merch_long', y='merch_lat', data=generated, ax=axes[1,1], alpha=0.1, color='green')
    axes[1,1].set_title('Generated Merchant Locations')
    
    plt.tight_layout()
    plt.show()

analyze_geography(original_data, generated_data)

## 5. Transaction Type Analysis

def analyze_transaction_types(original, generated):
    if 'transaction_type' not in original.columns or 'transaction_type' not in generated.columns:
        print("'transaction_type' column not found in both datasets")
        return
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # Original transaction types
    type_counts_orig = original['transaction_type'].value_counts(normalize=True)
    type_counts_orig.plot(kind='bar', ax=axes[0], color='blue')
    axes[0].set_title('Original Transaction Types')
    axes[0].set_xlabel('Transaction Type')
    axes[0].set_ylabel('Proportion')
    
    # Generated transaction types
    type_counts_gen = generated['transaction_type'].value_counts(normalize=True)
    type_counts_gen.plot(kind='bar', ax=axes[1], color='green')
    axes[1].set_title('Generated Transaction Types')
    axes[1].set_xlabel('Transaction Type')
    axes[1].set_ylabel('Proportion')
    
    plt.tight_layout()
    plt.show()

analyze_transaction_types(original_data, generated_data)

## 6. Temporal Pattern Analysis

def analyze_temporal_patterns(original, generated):
    if 'trans_date_trans_time' not in original.columns or 'trans_date_trans_time' not in generated.columns:
        print("'trans_date_trans_time' column not found in both datasets")
        return
    
    # Convert to datetime
    original['datetime'] = pd.to_datetime(original['trans_date_trans_time'])
    generated['datetime'] = pd.to_datetime(generated['trans_date_trans_time'])
    
    # Extract temporal features
    for df in [original, generated]:
        df['hour'] = df['datetime'].dt.hour
        df['day_of_week'] = df['datetime'].dt.dayofweek
        df['day_of_month'] = df['datetime'].dt.day
        df['month'] = df['datetime'].dt.month
    
    # Plot comparisons
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Hourly distribution
    sns.countplot(x='hour', data=original, ax=axes[0,0], color='blue')
    axes[0,0].set_title('Original - Hourly Distribution')
    sns.countplot(x='hour', data=generated, ax=axes[0,1], color='green')
    axes[0,1].set_title('Generated - Hourly Distribution')
    
    # Day of week distribution
    sns.countplot(x='day_of_week', data=original, ax=axes[1,0], color='blue')
    axes[1,0].set_title('Original - Day of Week Distribution')
    sns.countplot(x='day_of_week', data=generated, ax=axes[1,1], color='green')
    axes[1,1].set_title('Generated - Day of Week Distribution')
    
    plt.tight_layout()
    plt.show()

analyze_temporal_patterns(original_data, generated_data)

print("\nEnhanced EDA Complete! Review the visualizations and statistics above.")
```
